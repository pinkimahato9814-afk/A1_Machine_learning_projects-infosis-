```python
text = ' pinki123@gmail.com,ritu56@123,pinki#@#@6789 , 1234666,#contact me:345678.hello hello hello ..............'
```

```python
clean_space = re.sub(f'\s','',text)
clean_space
```

```python
remove_num = re.sub(r'\d+',' ',clean_space)
remove_num
```

```python
remove_ = re.sub(r'[,.:@#]','',remove_num)
remove_
```

```python
capital = re.sub(r'[^A-Za-z\s+]', '', text)
capital
```

```python
email = re.findall(r'\w+\d*@\w+.\w+', text)
email
```

```python
numbers = re.findall(r'\d+', text)

numbers
 
```
