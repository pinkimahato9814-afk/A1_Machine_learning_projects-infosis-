```python
# regular expression
import re
```

```python
'''
text = 'jfkdlajfklsd@#$78897u80'

re.sub(r'pattern', 'number', variable) -> remove

re.findall(r'pattern', variable) -> extract

str.replace(find, value, regex=True)

+ -> One or more

* -> Zero or more

'''
```

```python
text = "abc123@gmail.com  def@gmail.com  ghi@gmail.com    1234567. #Contact's Me: 34567890. Hello Hello Hello     .... "
```

```python
clean_space = re.sub(r'\s','',text)
clean_space
```

```python
clean_space = re.sub(r'\s+', ' ', text)
clean_space
```

```python
one_number = re.sub(r'\d+', '', clean_space)
one_number
```

```python
# remove dots '.'
remove_dot = re.sub(r'\.', '', one_number)
remove_dot
```

```python
remove_num = re.sub(r'\d+', '', clean_space)
remove_num
```

```python
# ^ -> not
remove_ = re.sub(r'[^.:@#]','',remove_num)
remove_
```

```python
capital = re.sub(r'[^A-Za-z\s+]', '', text)
capital
```
